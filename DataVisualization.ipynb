```python
import pandas as pd
df = pd.read_csv('/content/sample_data/student_habits_performance.csv')
print(df.head())
```

# **Distribution of Exam Scores**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['exam_score'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Exam Scores")
plt.xlabel("Score")
plt.ylabel("Student Count")
plt.grid(True)
plt.show()
```

# **Study Hours vs Exam Score**


```python
plt.figure(figsize=(10,6))
sns.regplot(x='study_hours_per_day', y='exam_score', data=df, scatter_kws={"alpha":0.4}, line_kws={"color":"red"})
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours Per Day")
plt.ylabel("Exam Score")
plt.show()
```

# Sleep Hours vs Exam Score

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='sleep_hours', y='exam_score', hue='diet_quality', data=df, palette='coolwarm')
plt.title("Sleep Hours vs Exam Score (colored by Diet Quality)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='sleep_hours', y='exam_score', hue='mental_health_rating', palette='coolwarm', data=df)
plt.title(" Sleep Hours vs Exam Score, Colored by Mental Health")
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Score")
plt.show()
```
